```python
fig = px.histogram(df, x= 'Distance_km', nbins=100,color_discrete_sequence=['black'])
fig2 = px.box(df, x= 'Distance_km',color_discrete_sequence=['black'])
fig.show()
fig2.show()
```

```python
fig3 = px.scatter(df , x= 'Preparation_Time_min', y = 'Delivery_Time_min'
                 ,color_discrete_sequence=['black'],trendline='ols',
                 trendline_color_override='red',template='simple_white')
fig3.show()
```

```python
fig = px.box(df ,x= 'Preparation_Time_min', y = 'Delivery_Time_min',color_discrete_sequence=['black']
            ,width =1200,template='simple_white')
fig.show()
```
